```python
import requests
import json

url = "https://api.yelp.com/v3/businesses/search"

API_KEY = "m_ptw8FCNofuzz24-_QHrWHQwE5bJsKQa_H5m8r3s_aTz8DN87q4z6y6w9ent9YvUGSHUMU_nNi0k6HZzAkNFP6toSzS6DxtVnbFO7s5133Wucw39fUfNT_pfpBtaHYx"


headers = {
    "Authorization": f"Bearer {API_KEY}",
    "Accept": "application/json"
}

params = {
    "location": "Bentonville",
    "term": "restaurants",
    "radius": 1000,
    "sort_by": "rating",
    "limit": 10
}

response = requests.get(url, headers=headers, params=params)

if response.status_code == 200:
    data = response.json()
    businesses = data.get("businesses", [])

    if not businesses:
        print("No restaurants found.")
    else:
        for idx, business in enumerate(businesses, start=1):
            name = business.get("name")
            rating = business.get("rating")
            address = ", ".join(business.get("location", {}).get("display_address", []))
            
            print(f"\nRestaurant #{idx}")
            print(f"Name: {name}")
            print(f"Rating: {rating}")
            print(f"Address: {address}")
else:
    print(f"Error {response.status_code}: {response.text}")
```
